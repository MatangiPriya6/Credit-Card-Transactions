```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset 
df = pd.read_csv('Credit card transactions - India - Simple.csv')
print(df.head())

```

```python
# Calculate the average transaction amount by city, card type, expenditure type, and gender
avg_amount_by_city = df.groupby('City')['Amount'].mean()
avg_amount_by_card_type = df.groupby('Card Type')['Amount'].mean()
avg_amount_by_exp_type = df.groupby('Exp Type')['Amount'].mean()
avg_amount_by_gender = df.groupby('Gender')['Amount'].mean()

```

```python
try:
    top_n = int(input("Enter the number of top cities to display: "))
except ValueError:
    print("Invalid input. Please enter a valid number.")
    exit()
```

```python
# Display top 10 cities
top_n_cities = avg_amount_by_city.nlargest(top_n)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_n_cities.values, y=top_n_cities.index, orient='h')
plt.title(f'Top {top_n} Cities by Average Transaction Amount')
plt.xlabel('Average Transaction Amount')
plt.ylabel('City')
plt.show()
```

```python

# Average amount by City
plt.subplot(1, 1, 1)
sns.barplot(x=avg_amount_by_city.index, y=avg_amount_by_city.values)
plt.title('Average Transaction Amount by City')
```

```python

# Average amount by card type
plt.subplot(1, 1, 1)
sns.barplot(x=avg_amount_by_card_type.index, y=avg_amount_by_card_type.values)
plt.title('Average Transaction Amount by Card Type')
```

```python
# Average amount by expenditure type
plt.subplot(1, 1, 1)
sns.barplot(x=avg_amount_by_exp_type.index, y=avg_amount_by_exp_type.values)
plt.title('Average Transaction Amount by Expenditure Type')

```

```python
# Average amount by gender
plt.subplot(2, 2, 3)
sns.barplot(x=avg_amount_by_gender.index, y=avg_amount_by_gender.values)
plt.title('Average Transaction Amount by Gender')
```
